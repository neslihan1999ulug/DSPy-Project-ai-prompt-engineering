```python
import dspy
from google.colab import userdata
import os

api_key = userdata.get("GEMINI_API_KEY")
os.environ["GEMINI_API_KEY"] = api_key

lm = dspy.LM("gemini/gemini-2.5-flash", api_key=api_key)
dspy.configure(lm=lm)

print("Bağlantı tamam!")
```

```python
class DuyguAnalizi(dspy.Signature):
    """Metnin duygusunu belirle."""
    metin: str = dspy.InputField()
    duygu: str = dspy.OutputField(desc="pozitif/negatif/nötr")
    neden: str = dspy.OutputField()

analiz = dspy.ChainOfThought(DuyguAnalizi)



sonuc = analiz(metin="Bugün harika bir gün geçirdim!")
sonuc1 = analiz(metin="Bugün harika bir gün!")
sonuc2 = analiz(metin="Çok kötü bir gün geçirdim.")
sonuc3 = analiz(metin="Sıradan bir günüm oldu.")

print("Duygu:", sonuc.duygu)
print("Neden:", sonuc.neden)
```
